```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set(style='darkgrid', color_codes=True)

df_ewm = pd.read_csv('df_ewm.csv', parse_dates=True)
df_ewm.set_index('timestamp').pivot(columns='alpha', values='ewma').plot(subplots=True)
```
